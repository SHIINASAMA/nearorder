## Disorder Metrics

| Symbol | Description |
| --- | --- |
| n | Number of elements |
| inversion\_count | Total number of inversions, used to measure global disorder; 0 indicates a fully ordered sequence |
| inversion\_ratio | Global disorder ratio |
| local\_inversion\_ratio | Ratio of adjacent inversions |
| max\_monotonic\_run | Length of the longest contiguous monotonic run |
| displacement\_sum | Average displacement of each element from its position in the corresponding sorted sequence |


```python
import json

from nearorder.math.generate import *
from nearorder.math.metrics import disorder_metrics

import matplotlib.pyplot as plt

seed = 114

target = base_sequence(100, "asc")
target = inject_adjacent_swaps(target, 10, seed=seed)
target = break_runs(target, 10)
target = partial_shuffle(target, 0.05, seed=seed)

print(target)

print(json.dumps(disorder_metrics(target), indent=2))

x = range(len(target))

plt.figure()
plt.plot(x, target)
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Target Sequence (Original Order)")
plt.grid(True)
plt.show()
```
